# Pairplot

```python
# pair plot eita individual datar stahe tar characterictics dekhai
# feature gula eke opor er sathe kivabe linkage eita dekhai
penguins=sns.load_dataset("penguins")

penguins.head()
```

```python
sns.pairplot(penguins, kind ="kde")
```

```python
sns.pairplot(data=penguins, hue="island")
```

```python
sns.pairplot(data=penguins, hue="species",markers=["o","s","d"],corner=True)
```

```python
sns.pairplot(data=penguins, hue="species", diag_kind="hist")
```

```python
sns.pairplot(data=penguins, hue="species", kind="kde")
```

# Heat Map

```python
## heat map

glue=sns.load_dataset("glue")
glue.head()
```

```python
mode_glue=sns.load_dataset("glue").pivot(index="Model",columns="Task",values="Score")

mode_glue.head(10)
```

```python
## amra color code dia bujaite chai kun method ta besi upojodi
### kintu eikhn a kun sell er man koto seita dekte passina
#### eijonno amader annot= True kore dilei amader level gula o dekhabe
sns.heatmap(mode_glue)
```

```python
sns.heatmap(mode_glue ,annot=True)  
## annodation true kore dile value gula o asbe
```

```python
## eikhn a axis ta amader change korte chaile ax.xaxis.tick_top() function dile
# axis opor a uthe jabe

ax= sns.heatmap(mode_glue ,annot=True)
ax.xaxis.tick_top()
```
