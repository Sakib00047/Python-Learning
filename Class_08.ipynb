```python
## graph e equation toiri kora

import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 5, 10)

# Constants
A1, A2 = 5, 2
a1, a2 = 2, 0.5  # Smaller a2 for better visualization

# Exponential functions
y1 = A1 * np.exp(-a1 * t)  # Decay
y2 = A2 * np.exp(a2 * t)   # Growth

# Check lengths
print(len(t), len(y1), len(y2))

# Plot the graphs
plt.plot(t, y1, color="r", label="y1: $A_1 e^{-a_1 t}$")
plt.plot(t, y2, color="b", label="y2: $A_2 e^{a_2 t}$")

# Add labels, title, and legend
plt.xlabel("Time")
plt.ylabel("Magnitude of the function")
plt.title("Exponential Curve of $\\alpha$ ($A e^{\\pm at}$)")
plt.legend()
plt.grid(True)  # Optional: Add grid for clarity
plt.show()
```

```python
# making Subplot 

import numpy as np
import matplotlib.pyplot as plt

# Sample data for plotting
t = np.linspace(0, 5, 10)
y1 = np.exp(-t)
x = np.random.rand(50)
y = np.random.rand(50)
value = [5, 10, 15, 20]  # Positions for the bars
height = [3, 7, 2, 5]    # Heights of the bars
mylabels = ['A', 'B', 'C', 'D']

plt.figure(figsize=(15, 10))

# First subplot - Bar chart
plt.subplot(2, 2, 1)
plt.bar(value, height, width=0.5)  # Adding height for the bars
plt.title("Bar Chart")

# Second subplot - Pie chart
plt.subplot(2, 2, 2)
plt.pie([15, 25, 35, 25], labels=mylabels, startangle=90, labeldistance=1.1, explode=(0.05, 0.05, 0.05, 0.05), colors=("r", "b", "m", "y"))
plt.title("Pie Chart")

# Third subplot - Line plot
plt.subplot(2, 2, 3)
plt.plot(t, y1, color="green", linewidth=4, linestyle="dashed", marker="d", markerfacecolor="blue", label="Rainfall")
plt.legend()
plt.title("Line Plot")

# Fourth subplot - Scatter plot
plt.subplot(2, 2, 4)
plt.scatter(x, y, color='r', marker='o')
plt.title("Scatter Plot")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
